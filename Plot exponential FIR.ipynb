```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
tau=2    # time constant
Ts=0.05 # sample rate (s)
x = [i*Ts for i in range(0,100)]
```

```python
y = [tau*np.exp(-tau*x[i]) for i in range(0,100)]  # a*exp(-a*x) = normalised!
```

```python
plt.plot(x,y)
```
